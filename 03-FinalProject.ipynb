## Background and Prior Work

We are trying to determine if we can find a relationship between a movie cast's stardom level and the movie's performance. Films packed with celebrities often can gain a lot of initial interest in trailers and marketing as there are people that show up at movie theaters just to see a specific celebrity on the screen. However, factors like negative online ratings and word of mouth reactions can tank a movie's performance a lot more than any celebrity can boost it, so it is hard to determine for any given movie how well it will perform. We will look for long term trends to discover if celebrity status is a good enough sign of movie success or not. We will define ways to measure stardom and measure movie success.

In "A Quantitative Study on the Effect of Cast Member Celebrity on Film Success," researchers used a regression model and analyzed the top 20 grossing films from 2015-2019, so a total of 100 movies total. They discovered that increased celebrity status correlated with both increased revenue and improved public perception. We question the conclusion of this article as taking the top 20 grossing films can mislead analysis as it doesn't account for cases where a movie "flops" but has celebrities. The author notes constraints such as using only the top 7 actors/actresses in their stardom metrics. Moreover, the study uses the actor/actress's current stardom level, which does not work in cases where the movie makes the actor/actress famous post-release. We can address these issues by accounting for a larger amount of the cast in stardom and measuring the actor/actress's popularity at the time of movie release. Additionally, 100 movies is likely not a large enough sample size and these movies were pulled up until 2019; Now a lot of marketing is done through social media which was not such a big tactic back then. Therefore, we can try to acquire data from a greater amount of movies and also use more recent movies.

The study finds that, on average, attaching a star adds a few million dollars to a movie’s expected box‑office, quantifying star power to 3‑million‑dollar increase in revenue rather than treating it as a vague marketing benefit. Elberse shows that this effect is not constant. The marginal impact of a new star is larger when the existing cast is already strong, meaning other members within groups are important rather than simple “one big name is enough” logic. By focusing on expected revenue before release, the paper frames stars as a costly but quantifiable investment whose value must be weighed against high salaries and profit participation. ### T-Test Analysis: IMDB Ratings by Star Cast Strength and Genre

To investigate whether star cast strength has a statistically significant association with IMDB ratings across different genres, we can perform genre-stratified independent samples t-tests. We test for normality in each group using a normality test, then compare IMDB ratings between movies with high vs. low star cast strength within each genre using t-tests. This analysis controls for genre as a potential confounding variable.

```python
# parse by genre
df_genres = df[['movie_name', 'Star_Cast_Strength', 'IMDB_Rating', 'Genre']].copy()
df_genres['Genre'] = df_genres['Genre'].str.split(r',\s*')
df_genres = df_genres.explode('Genre')

# get the top genres (>= 20 movies) for decent sample size
genre_counts = df_genres['Genre'].value_counts()
major_genres = genre_counts[genre_counts >= 20].index.tolist()

print(f"Analyzing {len(major_genres)} major genres with >= 20 movies each:")
print(f"{major_genres}")
print(f"\nGenre distribution:\n{genre_counts[genre_counts >= 20]}")
```

```python
from scipy.stats import normaltest, ttest_ind

ttest_results = []

for genre in major_genres:
    genre_data = df_genres[df_genres['Genre'] == genre][['Star_Cast_Strength', 'IMDB_Rating']].dropna()
    
    median_strength = genre_data['Star_Cast_Strength'].median()
    high_cast = genre_data[genre_data['Star_Cast_Strength'] >= median_strength]['IMDB_Rating']
    low_cast = genre_data[genre_data['Star_Cast_Strength'] < median_strength]['IMDB_Rating']
    
    if len(high_cast) > 0 and len(low_cast) > 0:
        # test normality for both groups
        high_stat, high_p_norm = normaltest(high_cast)
        low_stat, low_p_norm = normaltest(low_cast)
        
        # perform independent samples t-test with equal_var=True
        t_stat, p_value = ttest_ind(high_cast, low_cast, equal_var=True)
        
        ttest_results.append({
            'Genre': genre,
            'High_Cast_Mean': high_cast.mean(),
            'Low_Cast_Mean': low_cast.mean(),
            'Mean_Diff': high_cast.mean() - low_cast.mean(),
            'High_Normality_p': high_p_norm,
            'Low_Normality_p': low_p_norm,
            'T_Statistic': t_stat,
            'P_Value': p_value,
            'High_N': len(high_cast),
            'Low_N': len(low_cast),
            'Significant': 'Yes' if p_value < 0.05 else 'No'
        })

ttest_df = pd.DataFrame(ttest_results).sort_values('P_Value')

print("\nNormality Tests:")
print("-"*100)
print(ttest_df[['Genre', 'High_Normality_p', 'Low_Normality_p']].round(4).to_string(index=False))

```

We can see that most p values are >= 0.05 which signifies normal distribution, while there are a few outliers like Crime and Drama's High_Normality_p. These are not really an issue yet as this alone is not enough to draw conclusion on the data. Additionally, we chose these genres since they had >= 20 datapoints, and we have 52 and 189 datapoints for Crime and Drama genres, respectively, so our t-test should be robust to these normality violations.

```python
print("* p < 0.05 suggests data may not be normally distributed")
print("\nIndependent Samples T-Test: IMDB Rating (High vs Low Star Cast Strength)")
print("="*100)
print(ttest_df[['Genre', 'High_Cast_Mean', 'Low_Cast_Mean', 'Mean_Diff', 'T_Statistic', 'P_Value', 'Significant']].to_string(index=False))
print("\n* Significant at p < 0.05 level")
```

```python
# plot mean IMDB ratings for high vs low cast strength
fig1, ax1 = plt.subplots(figsize=(10, 6))

ttest_sorted = ttest_df.sort_values('Mean_Diff', ascending=True)
x_pos = np.arange(len(ttest_sorted))
width = 0.35

ax1.barh(x_pos - width/2, ttest_sorted['Low_Cast_Mean'], width, label='Low Cast Strength', alpha=0.7)
ax1.barh(x_pos + width/2, ttest_sorted['High_Cast_Mean'], width, label='High Cast Strength', alpha=0.7)
ax1.set_yticks(x_pos)
ax1.set_yticklabels(ttest_sorted['Genre'])
ax1.set_xlabel('Mean IMDB Rating', fontsize=11)
ax1.set_title('IMDB Ratings: High vs Low Star Cast Strength by Genre', fontsize=12)
ax1.legend()
ax1.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

As we can see, there really is not much difference between mean IMDB rating when comparing cast strength levels.

### T-Test Analysis: IMDB Ratings by Star Cast Strength and Year

We wanted to investigate and see wheteher cast strength had a statsitical impact on the ratings that each movie received by year. To do this, we compared movies with high star cast strenghts to movies with low star cast strength using t-tests. While working the analysis, we came across a few obstacles because of the lack of data after splitting and that will be explained in the discussion

```python
#ttest by year versus IMDB rating now 
# parse by year
df_year = df[['movie_name', 'Star_Cast_Strength', 'IMDB_Rating', 'Released_Year']].copy()

# get the top years (>= 10 movies) for decent sample size
year_counts = df_year['Released_Year'].value_counts()
major_years = year_counts[year_counts >= 10].index.tolist()

print(f"Analyzing {len(major_years)} major years with >= 10 movies each:")
print(f"{major_years}")
print(f"\nYear distribution:\n{year_counts[year_counts >= 10]}")
```

```python
ttest_results = []

for year in major_years:
    year_data = df_year[df_year['Released_Year'] == year][['Star_Cast_Strength', 'IMDB_Rating']].dropna()

    median_strength = year_data['Star_Cast_Strength'].median()
    high_cast = year_data[year_data['Star_Cast_Strength'] >= median_strength]['IMDB_Rating']
    low_cast = year_data[year_data['Star_Cast_Strength'] < median_strength]['IMDB_Rating']

    if len(high_cast) > 1 and len(low_cast) > 1:
        # only run normality test if sample is large enough
        high_p_norm = normaltest(high_cast).pvalue if len(high_cast) >= 8 else np.nan
        low_p_norm = normaltest(low_cast).pvalue if len(low_cast) >= 8 else np.nan

        t_stat, p_value = stats.ttest_ind(high_cast, low_cast, equal_var=True)

        ttest_results.append({
            'Released_Year': year,
            'High_Cast_Mean': high_cast.mean(),
            'Low_Cast_Mean': low_cast.mean(),
            'Mean_Diff': high_cast.mean() - low_cast.mean(),
            'High_Normality_p': high_p_norm,
            'Low_Normality_p': low_p_norm,
            'T_Statistic': t_stat,
            'P_Value': p_value,
            'High_N': len(high_cast),
            'Low_N': len(low_cast),
            'Significant': 'Yes' if p_value < 0.05 else 'No'
        })

ttest_df = pd.DataFrame(ttest_results).sort_values('Released_Year')
print("\nNormality Tests:")
print("-"*100)
print(ttest_df[['Released_Year', 'High_Normality_p', 'Low_Normality_p']].round(4).to_string(index=False))
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=ttest_df, x='Released_Year', y='Mean_Diff', marker='o')
plt.axhline(0, linestyle='--', color='red')
plt.title('Effect of Star Cast Strength on Rating Over Time')
plt.ylabel('Mean Difference (High - Low)')
plt.xlabel('Year')
plt.grid(alpha=0.3)
plt.xticks(ttest_df['Released_Year'], rotation=45)
plt.show()
```

As we can see here, it is evident that there is not a true strong correlation to cast strength and IMDB rating by year. The graph is showing the mean difference between high and low cast ratings. We get the high cast movies average ratings per year, and do the same with the low. Then from there, we subtract to get the mean difference and plot on the graph. We do this per year.

```python
def compute_year_corr(group):
    if len(group) < 10:
        return np.nan
    return group['Star_Cast_Strength'].corr(group['IMDB_Rating'])
year_corr = df_year.groupby('Released_Year').apply(compute_year_corr).reset_index(name='Correlation')
year_corr = year_corr.dropna()

plt.figure(figsize=(10,5))
sns.lineplot(data=year_corr, x='Released_Year', y='Correlation', marker='o')
plt.axhline(0, linestyle='--', color='green')
plt.title('Correlation Between Cast Strength and IMDB Rating by Year')
plt.ylabel('Correlation')
plt.xlabel('Year')
plt.xticks(year_corr['Released_Year'], rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

Here, we created a correlation graph between the star cast strenght and IMDb rating by year. In some years, we see a very high correlation between the strength and the rating, but other years siginifcantly less. ### Regression Analysis: How much does cast strength affect IMDb rating?

For the 2nd analysis we try using simple linear regression to test whether Star_Cast_Strength can predict IMDB_Rating. In our first analysis we used the t-test in which helps us answer whether the ratings are different between low or high cast strength groups. Regression ends up asking the question how cast increase in strength affects our IMDb rating.

After doing an initial run of linear regression in general, we also try regression separately for each major genre. By doing this we can kind of see if certain genres skew the trend or if these patterns are consistent among all genres. 

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

reg_df = df[['IMDB_Rating', 'Star_Cast_Strength', 'Released_Year', 'Genre']].dropna().copy()
reg_df['Released_Year'] = pd.to_numeric(reg_df['Released_Year'], errors='coerce')
reg_df = reg_df.dropna()

slope, intercept, r_value, p_value, std_err = stats.linregress(
    reg_df['Star_Cast_Strength'], reg_df['IMDB_Rating']
)

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R²: {r_value**2:.4f}")
print(f"P-value: {p_value:.4g}")
```

An initial analysis of these numbers shows us that the direction is positive since the slope is posive meaning stronger casts do have higher ratings. The R^2 value is the coefficient of determination meaning in this case how much IMDb ratings are explained by cast strength. Our low value suggests that not a significant amount of the differences in IMDb ratings can be explained by cast strength alone. However the P value is less than 0.05 meaning it is statistically signifcant. This means that the effect does exsit and is real but it's a tiny affect as indicated by the R^2.

```python
fig, ax = plt.subplots(figsize=(9, 5))
sns.regplot(data=reg_df, x='Star_Cast_Strength', y='IMDB_Rating',
            scatter_kws={'alpha': 0.4, 's': 20}, line_kws={'color': 'red'}, ax=ax)
ax.set_title('Star Cast Strength vs IMDb Rating (with regression line)')
ax.set_xlabel('Star Cast Strength')
ax.set_ylabel('IMDb Rating')
plt.tight_layout()
plt.savefig('results/regression_imdb_scatter.png', dpi=300)
plt.show()
```

```python
genres_regression = reg_df.copy()
genres_regression['Genre'] = genres_regression['Genre'].str.split(r',\s*')
genres_regression = genres_regression.explode('Genre')

genre_counts = genres_regression['Genre'].value_counts()
major_genres = genre_counts[genre_counts >= 20].index.tolist()

genre_results = []
for genre in major_genres:
    g = genres_regression[genres_regression['Genre'] == genre].dropna()
    if len(g) > 5:
        s, i, r, p, se = stats.linregress(g['Star_Cast_Strength'], g['IMDB_Rating'])
        genre_results.append({'Genre': genre, 'Slope': s, 'R2': r**2, 'P_Value': p})

genre_reg_df = pd.DataFrame(genre_results).sort_values('Slope')

fig, ax = plt.subplots(figsize=(10, 6))
colors = ['green' if p < 0.05 else 'steelblue' for p in genre_reg_df['P_Value']]
ax.barh(genre_reg_df['Genre'], genre_reg_df['Slope'], color=colors, alpha=0.75)
ax.axvline(0, color='red', linestyle='--', linewidth=1.5)
ax.set_xlabel('Regression Slope')
ax.set_title('Regression Slope by Genre\n(Green = significant and p < 0.05)')
plt.tight_layout()
plt.savefig('results/regression_imdb_by_genre.png', dpi=300)
plt.show()
```

We can kind of see from the graph that most of the genres have a positive slope so there must be a higher cast strength and that means they do correlate withe higher ratings in those genres. However, it may be a weak correlation. The only significant correlations that are statistically significant are action and drama and the rest are not very highly correlated or aren't as significant. Also, generes like fantasy, comedy, and also romance are negative so that would mean that in those cases stronger casts end up being associated wiht lower IMDb ratings. 

This actually matches up what we saw in the T-Test. 

A possible reason for this could be that in certain types of stories and movies like Romance or Fantasy it's not necessarily about the cast strength but perhaps other factors play a larger role in the movie's success and ratings. 

```python
reg_df['Predicted'] = intercept + slope * reg_df['Star_Cast_Strength']
reg_df['Residual'] = reg_df['IMDB_Rating'] - reg_df['Predicted']

fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(reg_df['Predicted'], reg_df['Residual'], alpha=0.4, s=20)
ax.axhline(0, color='red', linestyle='--')
ax.set_xlabel('Predicted IMDb Rating')
ax.set_ylabel('Residual from actual rating')
ax.set_title('Residuals vs Actual IMDb Rating)')
plt.tight_layout()
plt.savefig('results/regression_imdb_residuals.png', dpi=300)
plt.show()
```
