```python
top_1000.isnull()
```

We felt that poster link, certificate, and overview were not very relevant to our research question. We are measuring the relationship between the strength of the star cast and movie rating/revenue. The poster link and certificate would not be of much help, and not many people watch movies because of the overview, which we would also need to transform into vectors or embeddings to determine whether it could be a confounding variable.

```python
# remove irrelevant columns, then drop nulls
top_1000 = top_1000[['Series_Title', 'Released_Year', 'Runtime', 'Genre', 'IMDB_Rating', 'Meta_score', 'Director', 'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes', 'Gross']]
top_1000.to_csv('data/01-interim/imdb_top_1000_relevant_cols.csv')
rows_to_drop = top_1000[top_1000.isnull().any(axis=1)]
rows_to_drop
```

```python
# clean dataframe by dropping all Nan rows
top_1000 = top_1000.dropna()
top_1000.to_csv('data/01-interim/imdb_top_1000_dropna.csv')
top_1000.isnull().any(axis=1)
```

```python
top_1000.isnull().any(axis=1).any()
```
