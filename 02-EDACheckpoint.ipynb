# Research Question

Is there a statistically significant relationship between the prestige of a film cast, measured by major award nominations and wins, and how the film is rated by viewers, measured by IMDb ratings, controlling for genre and release year?


## Background and Prior Work

We are trying to determine if we can find a relationship between a movie cast's stardom level and the movie's performance. Films packed with celebrities often can gain a lot of initial interest in trailers and marketing as there are people that show up at movie theaters just to see a specific celebrity on the screen. However, factors like negative online ratings and word of mouth reactions can tank a movie's performance a lot more than any celebrity can boost it, so it is hard to determine for any given movie how well it will perform. We will look for long term trends to discover if celebrity status is a good enough sign of movie success or not. We will define ways to measure stardom and measure movie success.

In "A Quantitative Study on the Effect of Cast Member Celebrity on Film Success," researchers used a regression model and analyzed the top 20 grossing films from 2015-2019, so a total of 100 movies total. They discovered that increased celebrity status correlated with both increased revenue and improved public perception. We question the conclusion of this article as taking the top 20 grossing films can mislead analysis as it doesn't account for cases where a movie "flops" but has celebrities. The author notes constraints such as using only the top 7 actors/actresses in their stardom metrics. Moreover, the study uses the actor/actress's current stardom level, which does not work in cases where the movie makes the actor/actress famous post-release. We can address these issues by accounting for a larger amount of the cast in stardom and measuring the actor/actress's popularity at the time of movie release. Additionally, 100 movies is likely not a large enough sample size and these movies were pulled up until 2019; Now a lot of marketing is done through social media which was not such a big tactic back then. Therefore, we can try to acquire data from a greater amount of movies and also use more recent movies.

Additionally, in "The Power of Stars: Do Star Actors Drive the Success of Movies?" the researchers examine whether hiring star actors measurably increases a film’s expected theatrical revenue, using prediction‑market data and stock‑market style reactions around casting announcements. The study finds that, on average, attaching a star adds a few million dollars to a movie’s expected box‑office, quantifying star power to 3‑million‑dollar increase in revenue rather than treating it as a vague marketing benefit. Elberse shows that this effect is not constant. The marginal impact of a new star is larger when the existing cast is already strong, meaning other members within groups are important rather than simple “one big name is enough” logic. By focusing on expected revenue before release, the paper frames stars as a costly but quantifiable investment whose value must be weighed against high salaries and profit participation. However, a "star" is defined as actors who already have HSX StarBonds, which requires at least two major roles. That excludes emerging talent and underestimates cases where a film itself creates a star which this paper ignores, a similar flaw to the one found in the first document.

In determining how we determine "prestige" of a movie's cast, we came across "Factors Affecting the Financial Success of Motion Pictures: What is the Role of Star Power?" in which they coin the term "star power" for this metric. In this paper they describe using the number of nominations for Academy awards and average lifetime gross revenue of films involving the specific celebrity. Again, to avoid previously discussed issues, we wlil have to make sure we use these stats respective to the time in wihch the movie came out, not present day. This document also reveals some surprising findings like that the time of year of the movie's release didn't seem to matter, nor did the genre make a significant impact. We will still take these into account when analyzing these movies, as we can never be completely sure if there truly is no effect from these features. They end up acknowledging how subjective star power is given how everyone has their unique perception and is thus difficult to measure. This only reinforces the notion that we should find many different ways to measure stardom.

We also found a similar project on Github, "IMDb_ratings_analysis" that goes more broad and tries to find multiple parameters for a movie's success, as well as predict if a movie is going to do well based on discovered factors. They even perform plot analysis as part of their analysis. We do not plan on going this far with our project, but these are good stretch goals to include. #### Most Frequently Occuring Actors

The first step of this exploratory analysis is to see how frequently certain actors and directors show up in the dataset, as well as who may be showing up the most. This can give us some insights into the more "popular" actors, which can allow us to determine whether actor popularity is correlated with movie rating or revenue. We do this by checking how often each actor appears in either the Star1, Star2, Star3, or Star4 column, as an actor can appear as a primary or secondary star in any movie.

```python
## YOUR CODE HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
movies = pd.read_csv('data/02-processed/movies_combined_cleaned.csv')
movies.head()
```

```python
# the following line was partially generated by ChatGPT using the prompt "get all unique values from multiple columns in a pandas df"
all_actors = movies[['Star1', 'Star2', 'Star3', 'Star4']].stack()

# the following two lines were not generated by AI
top_10_freqs = all_actors.value_counts()[:10]
top_10_actors = sns.barplot(x=top_10_freqs.values, y=top_10_freqs.index)
top_10_actors.set(xlabel='Frequencies', ylabel='Actor Names')
```

```python
# create a set of all actors, then check how many times they appear in either Star1, Star2, Star3, or Star4

actors = set(movies['Star1']) | set(movies['Star2']) | set(movies['Star3']) | set(movies['Star4'])
freqs = {}
star1_val_counts = movies['Star1'].value_counts(normalize=True)
star2_val_counts = movies['Star2'].value_counts(normalize=True)
star3_val_counts = movies['Star3'].value_counts(normalize=True)
star4_val_counts = movies['Star4'].value_counts(normalize=True)

for actor in actors:
  actor_col = f'{actor}_Freq'
  # get the frequency that each actor shows up in the dataframe
  freqs[actor_col] = (star1_val_counts.get(actor, 0) + star2_val_counts.get(actor, 0) + star3_val_counts.get(actor, 0) + star4_val_counts.get(actor, 0))

freqs
```

```python
len(actors)
```

```python
len(actors) == len(freqs)
```

```python
freqs['Christian Bale_Freq']
```

```python
np.mean(list(freqs.values()))
```

We now aggregate the frequencies into one metric for each movie by adding up each cast member's frequency

```python
# Add up all the frequencies of each star cast member to get the strength
def get_star_cast_strength(row):
  star1_name = f'{row['Star1']}_Freq'
  star2_name = f'{row['Star2']}_Freq'
  star3_name = f'{row['Star3']}_Freq'
  star4_name = f'{row['Star4']}_Freq'
  strength = freqs[star1_name] + freqs[star2_name] + freqs[star3_name] + freqs[star4_name]
  return strength

movies['Star_Cast_Strength'] = movies.apply(get_star_cast_strength, axis=1)
movies.head()
```

The above confirms that Christian Bale is one of the most mentioned actors in this dataset. On average, an actor appears 0.5% of the time in the dataset, but Christian Bale appeared about 3% of the time. This gives us a good idea of which actors overall show up the most in the dataset - but that also may not indicate overall popularity since we have a small sample size of 288 and some popular actors may simply not show up in this dataset. Regardless, this gives us a decent idea of which actors are generally more popular, and whether their movies rate or gross higher. We can now plot the average of our output metrics, such as IMDB rating, movie score, and gross revenue for the movie.

#### Average/Median IMDB Rating, Movie Score, and Gross Revenue

We decided to summarize a lot of our data through some indicators. We will find the mean and median. The mean is needed so we find out if there are any outliers since sometimes if a few really good movies have a good rating then we can see that. We also look for the median since it is a bit better to see what an average movie would be in terms of rating. If the data is skewed then the median would be a better predictor. So we end up computing the mean and median for IMDB rating, movie score, and the gross revenue which is overall revenue.

```python
key_vars = ['IMDB_Rating', 'Meta_score', 'Gross']
summary = movies[key_vars].agg(['mean', 'median', 'std', 'min', 'max'])
summary.round(2)
```

For **IMDB_Rating** and **Meta_score** the mean and median are pretty similar so they are symmetric almost. For **Gross** the mean is pretty higher so there must be a skew of some kind where a few high grossing films make the average go up.

```python
#used the help of chatgpt here for the syntax of .dropna()
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, col in zip(axes, key_vars):
    movies[col].dropna().hist(ax=ax, bins=25, edgecolor='black', alpha=0.7)
    ax.set_title(col.replace('_', ' '))
    ax.set_xlabel(col.replace('_', ' '))  # x-axis = variable values (rating, score, or revenue)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.suptitle('Distributions of IMDB Rating, Meta score, and Gross revenue', y=1.02, fontsize=12)
plt.show()
```

The histograms have these patterns:

- **IMDB_Rating and Meta_score** are similar to each other since a majority of the data is near the center and the left and right sides are very similar. When a distribution is symmetric we know that the mean and median are similar and close to each other.

- **Gross** is very **right skewed**. Even though most movies in the sample do an average amount of revenue a smaller but decent number of high grossing films create the right skew. This means that the mean is affected by values that are extreme and in that case the mean is usually larger than the median. SO the average isn't really representatitve of the actual distribution. The median though is where half the movies are below and half are above the value. 

Because of this, we will end up using median to describe Gross revenue so our conclusions aren't affected by outliers. And we can use either mean or median for the IMDB_Rating and Meta_score.
